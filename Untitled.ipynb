```python
import random
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import spacy
from utils import translate_sentence, bleu, save_checkpoint, load_checkpoint
from torch.utils.tensorboard import SummaryWriter  # to print to tensorboard
from torchtext.legacy.datasets import Multi30k
from torchtext.legacy.data import Field, BucketIterator
```

```python
import bn_core_news_sm
spacy_ben = bn_core_news_sm.load()
spacy_eng = spacy.load("en_core_web_sm")


def tokenize_ben(text):
    return [tok.text for tok in spacy_ben.tokenizer(text)]


def tokenize_eng(text):
    return [tok.text for tok in spacy_eng.tokenizer(text)]
```

```python

bangla = Field(tokenize=tokenize_ben, lower=True, init_token="<sos>", eos_token="<eos>")

english = Field(tokenize=tokenize_eng, lower=True, init_token="<sos>", eos_token="<eos>")
```

```python
train_data, valid_data, test_data = Multi30k.splits(
    exts=(".bn", ".en"), fields=(bangla, english),
    root='.data', train='corpus.train', validation='RisingNews.valid', test='RisingNews.test'
)

```

```python
bangla.build_vocab(train_data, max_size=10000, min_freq=2)
english.build_vocab(train_data, max_size=10000, min_freq=2)
```

```python
train_iterator, valid_iterator, test_iterator = BucketIterator.splits(
    (train_data, valid_data, test_data),
    batch_size=32,
    sort_within_batch=True,
    sort_key=lambda x: len(x.src),
    device=torch.device("cuda" if torch.cuda.is_available() else "cpu"),
)
```

```python


for batch_idx, batch in enumerate(train_iterator):
    inp_data = batch.src.to("cpu")
    target = batch.trg.to("cpu")
    bn_text_to_indices = [bangla.vocab.stoi[token] for token in inp_data]
    en_text_to_indices = [english.vocab.stoi[token] for token in target]

    # Convert to Tensor
    bn_sentence_tensor = torch.LongTensor(bn_text_to_indices).unsqueeze(1).to("cpu")
    en_sentence_tensor = torch.LongTensor(en_text_to_indices).unsqueeze(1).to("cpu")

    bn_translated_sentence = [bangla.vocab.itos[idx] for idx in bn_sentence_tensor]
    en_translated_sentence = [english.vocab.itos[idx] for idx in en_sentence_tensor]
        
    print(bn_translated_sentence, en_translated_sentence)
```
